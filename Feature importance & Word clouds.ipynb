```python
# Create a new DataFrame with just the "features" and "Netherlands" columns
#df_lsa
NL_lsa1 = EP6_lsa.loc[:,["words" ,"1"]]
NL_lsa0 = EP6_lsa.loc[:,["words" ,"0"]]# Extract the row index and the columns "01", "0", and "10" from the df_lsa DataFrame
NL_lsa10 = EP6_lsa.loc[:,["words" ,"10"]]
```

```python
NL_lsa0 = NL_lsa0.sort_values("0", ascending=False)
NL_lsa0 = NL_lsa0.iloc[:25]
NL_lsa1 = NL_lsa1.sort_values("1", ascending=False)
NL_lsa1 = NL_lsa1.iloc[:25]
NL_lsa10 = NL_lsa10.sort_values("10", ascending=False)
NL_lsa10 = NL_lsa10.iloc[:25]
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Get the words in the index of NL_lsa
words = list(NL_lsa0["words"]) #2nd most important

# Create a string of all the words
text = ' '.join(words)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=800, background_color='white').generate(text)

# Plot the word cloud
plt.figure(figsize=(8,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
