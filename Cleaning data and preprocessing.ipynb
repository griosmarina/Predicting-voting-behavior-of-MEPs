```python
import nltk
from nltk import word_tokenize, pos_tag
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
import nltk

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = nltk.corpus.stopwords.words('english')
#stop_words.extend(["more","words"])
stop_words.sort()
```

```python
dataset = EP7["raw_text"].to_list()
print(dataset)
```

```python
punctuation = [',', '.', ';', ':', '"', "-", "@", "#"
               '`', '´', "’", "'", "!", "?", "(", ")",
               "/", "$", "%"]


concat = []

wnl = WordNetLemmatizer()

rejected_words = {}
replacement_dict = {}
rejected_lists = {}

for txt in dataset:

  list_words = [wnl.lemmatize(i.lower(),j[0].lower()) if j[0].lower() in ['a','n','v'] else wnl.lemmatize(i.lower()) for i,j in pos_tag(word_tokenize(txt))]
  
  remaining_words  = []
  for word in list_words:
    
    # for punct in punctuation:
    #   word.replace(punct,"")

    word = word.lower()
    reason = ""
    if any(char.isdigit() for char in word): reason = "Number"
    #elif d.check(word)==False: reason = "Not a real word"
    elif word in stop_words: reason = "Stop word"
    elif len(word)<3: reason = "Too short"

    else:
      in_word = word
      out_word = stemmer.stem(in_word)
      
      remaining_words.append(out_word)
      replacement_dict.setdefault(out_word, set([])).update([in_word])

    if reason !="":
      rejected_words.setdefault(word, set([reason]))
      rejected_lists.setdefault(reason, set([])).update([word])

    concat.append(remaining_words)
```

```python
print("Number:",rejected_lists["Number"])
print("Not a real word:",rejected_lists["Not a real word"])
print("Stop word:",rejected_lists["Stop word"])
print("Too short:",rejected_lists["Too short"])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(lowercase=True, strip_accents='unicode', min_df=0.05, max_df=0.9, ngram_range=(1,2))
tfidf_array = tfidf_vectorizer.fit_transform(concat)
feature_names = tfidf_vectorizer.get_feature_names_out()
print(feature_names)
```

```python
from sklearn.decomposition import TruncatedSVD
number_dimensions = 4
lsa_model = TruncatedSVD(n_components=number_dimensions, random_state=420)
lsa_comps = lsa_model.fit_transform(tfidf_array)
print(lsa_comps)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(lsa_comps)
```
